```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
```

---

#### Link da documentação

- https://github.com/allisonhorst/palmerpenguins

---

```python
pinguins = sns.load_dataset("penguins")
pinguins.head()
```

# Distribution Plots

### Distplot
 - https://seaborn.pydata.org/generated/seaborn.distplot.html\

```python
sns.distplot(pinguins['bill_length_mm'], bins= 20).set(title='Distplot')
```

```python
sns.displot(pinguins['bill_length_mm'], bins = 20, kde = True).set(title='Displot')
```

```python
sns.histplot(pinguins['bill_length_mm'], bins = 20, kde = True).set(title='Histplot')
```

### Jointplot
 - https://seaborn.pydata.org/generated/seaborn.jointplot.html
 - https://matplotlib.org/3.3.3/api/markers_api.html

```python
sns.jointplot(data = pinguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
```

```python
sns.jointplot(data = pinguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species', kind = 'kde')
```

```python
sns.jointplot(data = pinguins, x = 'bill_length_mm', y = 'bill_depth_mm', kind="reg", marker = 'v')
```

```python
sns.jointplot(data=pinguins, x="bill_length_mm", y="bill_depth_mm", kind="hist", bins = 20, hue = 'species')
```

```python
sns.jointplot(data=pinguins, x = "bill_length_mm", y = "bill_depth_mm", kind = "hex")
```

### Barplot
 - https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
sns.barplot(data = pinguins, x = 'species', y = 'bill_length_mm')
```

### Countplot
 - https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
sns.countplot(data = pinguins, x = 'species').set(title = 'CountPlot')
```

```python
pinguins['species'].value_counts()
```

### Boxplot
 - https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
sns.boxplot(data = pinguins, x = 'species', y = 'body_mass_g', hue = 'sex')
```

### Violinplot
 - https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
sns.violinplot(data = pinguins, x = 'species', y = 'body_mass_g', hue = 'sex', split = True)
```

# Heatmap

 - https://seaborn.pydata.org/generated/seaborn.heatmap.html
 - https://matplotlib.org/3.5.1/gallery/color/colormap_reference.html

```python
sns.heatmap(pinguins.corr(), annot = True, cmap = 'cividis', linewidth = 2, linecolor = 'white')
```

### FacetGrid
 - https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

```python
pinguins.head()
```

```python
p = sns.FacetGrid(pinguins, col = 'island', row = 'species')
p.map(sns.histplot, 'body_mass_g', bins = 25, kde = True)
```

```python
p = sns.FacetGrid(pinguins, col = 'species', row = 'sex')
p.map(sns.histplot, 'bill_depth_mm', bins = 25, kde = True)
```
