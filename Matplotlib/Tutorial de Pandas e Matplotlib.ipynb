# Tutorial de Matplotlib

```python
# Criando um novo DataFrame e modificando a forma de visualizar as informações do DF anterior
carros = carros2.T
```

```python
carros
```

```python
# Criando uma coluna com valores acumulados de cada marca

for l in carros.columns:
    carros[f'{l} Acumulado'] = carros[l].cumsum()
```

```python
carros
```

# Funções

```python
# Plotando o primeiro gráfico

plt.plot(carros['Mustang Acumulado'], label = 'Vendas de Mustangs')
plt.title('Vendas: Mustang', fontsize=22)
plt.xlabel('Meses', fontsize=16)
plt.ylabel('Unidades Vendidas', fontsize=16)
plt.legend()
```

```python
# Criando subplots

plt.subplot(2, 2, 1)
plt.plot(carros['Ferrari'])
plt.xticks(rotation=90)

plt.subplot(2, 2, 2)
plt.plot(carros['Fiat'])
plt.xticks(rotation=90)

plt.subplot(2, 2, 3)
plt.plot(carros['Ford'])
plt.xticks(rotation=90)

plt.subplot(2, 2, 4)
plt.plot(carros['BMW'])
plt.xticks(rotation=90)


plt.tight_layout()
plt.show()
```

# Orientação a Objetos 

- https://matplotlib.org/3.5.1/gallery/lines_bars_and_markers/linestyles.html

- https://matplotlib.org/3.3.3/api/markers_api.html

```python
fig = plt.figure()



ax = fig.add_axes([0, 0, 1, 1])
ax.plot(carros['BMW Acumulado'], linestyle='--', color = 'orange', label = 'BMW', linewidth=4, alpha=0.7, marker='v',
       markersize=15, markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')
ax.plot(carros['Audi Acumulado'], linestyle='-.', color = 'navy', label = 'AUDI')
ax.set_title('Comparativo: Audi x BMW', fontsize=22)
ax.set_xlabel('Meses', fontsize=18,)
ax.set_ylabel('Unidades vendidas', fontsize=18)
plt.xticks(rotation=90)
ax.legend()


# -----------------------------------------------------------------------
# Adicionando o gráfico menor
# -----------------------------------------------------------------------


ax2 = fig.add_axes([0.625, 0.1, 0.3, 0.3])
ax2.plot(carros['BMW Acumulado'], linestyle='--', color = 'orange', label = 'BMW')
ax2.plot(carros['Audi Acumulado'], linestyle='-.', color = 'navy', label = 'AUDI')
ax2.set_xlim(['Mar','Jul'])
ax2.set_ylim([50000, 150000])
ax2.set_title('Março - Julho')
plt.xticks(rotation=90)


plt.show()
# Para salvar a figura
fig.savefig('AUDIxBMW.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(carros['BMW Acumulado'], linestyle='--', color = 'orange', label = 'BMW', linewidth=4, alpha=0.7, marker='v',
       markersize=15, markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')
ax.plot(carros['Audi Acumulado'], linestyle='-.', color = 'navy', label = 'AUDI')

ax.set_xlim(['Mar','Out'])
ax.set_ylim([50000, 150000])

ax.set_title('Comparativo: Audi x BMW', fontsize=22)
ax.set_xlabel('Meses', fontsize=18,)
ax.set_ylabel('Unidades vendidas', fontsize=18)
plt.xticks(rotation=90)
ax.legend(loc=2)
plt.show()
```
