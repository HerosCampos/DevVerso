### Juntando todas as marcas

```python
carros
```

```python
carros_original = pd.DataFrame(np.random.randint(3000, 25000, size=(8,12)), 
                   ['Fiat', 'Pegeout', 'Volkswagem', 'Ford', 'Nissan', 'Toyota', 'Honda', 'Chevrolet'], 
                   ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],)


carros_original['Total'] = carros_original.sum(axis=1)


carros_origem = ['Italia', 'França', 'Alemanha', 'EUA', 'Japão', 'Japão', 'Japão', 'EUA']
carros_original['Origem'] = carros_origem
```

```python
carros_original
```

### Concatenando os dois DataFrames

```python
carros_completo = pd.concat([carros, carros_original])
carros_completo
```

### Utilizando List Comprehensions e Lambda

```python
carros_completo['Sigla List Comprehensions'] = [l[:3].upper() for l in carros_completo['Origem']]
```

```python
carros_completo['Sigla Lambda Function'] = carros_completo['Origem'].apply(lambda x: x[:3].upper())
```

```python
carros_completo
```

### Trocando o nome das colunas
