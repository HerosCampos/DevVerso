```python
bitcoin = web.get_data_yahoo(['BTC-USD'], 
                          start=datetime.datetime(2018, 1, 1), 
                          end=datetime.datetime(2022, 4, 1))
```

```python
bitcoin
```

```python
bitcoin['Close'].plot(figsize = (16, 10))
```

```python
btc = bitcoin.loc['2021-01': '2022-04'].copy()
btc
```

### MÃ©dia MÃ³vel Simples

```python
btc['SMA30'] = btc['Close'].rolling(window = 30).mean()
btc
```

```python
fig = plt.figure(figsize = (14, 8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(btc['Close'], label = 'Bitcoin')
ax.plot(btc['SMA30'], label = 'SMA30')
ax.plot(btc['Close'].rolling(window = 15).mean(), label = 'SMA15')

ax.set_title('Bitcoin', fontsize = 20)
ax.set_ylabel('PreÃ§o ($)', fontsize = 16)

plt.legend()
plt.show()
```

---

### MÃ©dia MÃ³vel Exponencial

- https://pandas.pydata.org/pandas-docs/version/0.17.0/generated/pandas.ewma.html

- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.ewm.html

```python
btc['EMA30'] = btc['Close'].ewm(span = 30).mean()
btc
```

```python
fig = plt.figure(figsize = (14, 8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(btc['Close'], label = 'Bitcoin')
ax.plot(btc['EMA30'], label = 'EMA30', color = 'navy')
ax.plot(btc['Close'].ewm(span = 15).mean(), label = 'EMA15', color = 'orange')

ax.set_title('Bitcoin com mÃ©dias mÃ³veis ', fontsize = 20)
ax.set_ylabel('PreÃ§o ($)', fontsize = 16)

plt.legend()
plt.show()
```

### Visualizando as duas mÃ©dias juntas

```python
btc[['Close', 'EMA30', 'SMA30']].plot(figsize = (16,10))
```
